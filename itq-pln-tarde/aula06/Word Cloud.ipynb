```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
from wordcloud import WordCloud
```

```python
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv("../../../dados/nlp/analise_sentimento_portugues.csv", encoding="utf-8")
```

```python
df
```

```python
del df["sentimento"]
```

```python
texto_geral = ""
for texto in df["paragrafo"]:
    texto_geral = texto_geral + " " + texto
texto_geral
```

```python
tokens = word_tokenize(texto_geral.lower())
```

```python
stop_words = ["e", "o", "da", "de", "um", "uma", "na", "no", "á", "à"] 
```

```python
tokens_limpos = []
for token in tokens: 
    if token not in stop_words: 
        tokens_limpos.append( token )
texto_limpo = " ".join(tokens_limpos)
```

```python
wordcloud = WordCloud(width=1024, height=800, background_color='black').generate(texto_limpo)
```

```python
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
