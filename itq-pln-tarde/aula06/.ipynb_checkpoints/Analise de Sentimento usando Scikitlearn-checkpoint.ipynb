```python
stop_words = ["e", "o", "da", "de", "um", "uma", "na", "no", "á", "à"] 
```

```python
def limpar_texto( texto : str ) -> str:
    tokens = word_tokenize(texto.lower())
    tokens_limpos = []
    for token in tokens: 
        if token not in stop_words: 
            tokens_limpos.append( token )
    texto_limpo = " ".join(tokens_limpos)
    return texto_limpo
```

```python
df.loc[0]["paragrafo"]
```

```python
limpar_texto( df.loc[0]["paragrafo"] )
```

```python
df["paragrafo_limpo"] = df["paragrafo"].apply( limpar_texto )
```

```python
vetorizador = CountVectorizer()
bow = vetorizador.fit_transform( df["paragrafo"]  )
```

```python
bow.shape
```

```python
bow.toarray()
```

```python
vocabulario = vetorizador.get_feature_names_out()
vocabulario
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LogisticRegression()
```

```python
reg.fit(bow, df["sentimento"])
```

```python
reg.score(bow, df["sentimento"])
```

```python
vetorizador = CountVectorizer(max_features=165, lowercase=True, vocabulary=vocabulario)
```
