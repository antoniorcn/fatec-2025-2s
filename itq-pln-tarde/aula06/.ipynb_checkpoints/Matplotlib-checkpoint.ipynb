```python
x = np.arange(-5, 5, 0.10)
```

```python
xsin = np.sin(x)
```

```python
xsquare = x ** 2
```

```python
# plt.plot(x, xsin, label="Seno", color="green", marker="o", markerfacecolor="blue", markersize=5, linewidth=4, 
         # linestyle="dotted")
plt.barh(x, xsin, label="Seno", color="green")
plt.scatter(x, xsquare, label="Quadrado")
plt.ylabel("Seno de X")
plt.xlabel("Valores de X")
plt.text(-2, -1, "Vale do seno")
plt.ylim(-1, 1)
plt.xlim(-2, 2)
plt.title("Grafico de encontro do Seno com o Quadrado de X")
plt.legend()
plt.savefig("grafico1.png")
plt.show()
```
