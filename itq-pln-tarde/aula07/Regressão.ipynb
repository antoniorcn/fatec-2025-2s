```python
X = np.random.randint( 1, 50, 50) # Area plantada em Hectares
Y = np.random.randint(200, 400, 50) * X + 5000 # Toneladas colhidas
```

```python
X[4]
```

```python
Y[4]
```

```python
m = len(X)
```

```python
alpha = 0.00001
```

```python
w = 1000
b = 100
```

```python
# plt.scatter(X, Y)
# plt.plot(X, Y_hat, color="red")
# plt.show()
```

```python
for j in range(1000):
    soma_erros_dw = 0
    soma_erros_db = 0
    for i in range( m ):
        y_hat = w * X[i] + b
        erro = y_hat - Y[i]
        soma_erros_dw += (erro * X[i])
        soma_erros_db += erro
    w = w - (alpha * soma_erros_dw)
    b = b - (alpha * soma_erros_db)
    print(soma_erros_dw, w, b)
```

```python
Y_hat = w * X + b
plt.scatter(X, Y)
plt.plot(X, Y_hat, color="red")
plt.show()
```
