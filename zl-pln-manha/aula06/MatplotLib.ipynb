```python
vendas = [1000.0, 1300.0, 1500.0, 2100.0, 3200.0, 1700.0, 1300.0]
investimentos = [2000.0, 1000.0, 1500.0, 1200.0, 800.0, 900.0, 1000.0]
# tempo = [1, 2, 3, 4, 5, 6, 7]
tempo = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul" ]

plt.plot(tempo, vendas, color="green", marker="o", markersize = 10, markerfacecolor="darkgreen", label="Vendas")
plt.plot(tempo, investimentos, color="red", marker="o", markersize = 10, markerfacecolor="darkred", label="Investimentos")
plt.xlabel("Meses")
plt.ylabel("Valores")
plt.text(4, 3000, "Dia das mes")
plt.title("Vendas 1o semestre 2025")
plt.legend()
plt.savefig("./vendas.png")
plt.show()
```

```python
X = np.arange(-10, 10.25, 0.25)
```

```python
Y_NEG = -(X ** 2)
```

```python
Y = X ** 2
```

```python
plt.scatter(X, Y, linestyle="dashdot", linewidth=5, color="green")
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")
plt.xlim(-5, 5)
plt.ylim(-25, 25)
plt.axhline( 10 ) 
plt.axvline( 0 ) 
plt.plot(X, Y_NEG)
plt.show()
```
