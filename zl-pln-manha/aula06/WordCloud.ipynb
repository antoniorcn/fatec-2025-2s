```python
import matplotlib.pyplot as plt
```

```python
from wordcloud import WordCloud
```

```python
import pandas as pd
```

```python
df = pd.read_csv("../../../dados/nlp/analise_sentimento_portugues.csv", encoding="utf-8")
```

```python
texto = "" 
for paragrafo in df["paragrafo"]:
    texto += paragrafo + " "
texto
```

```python
stop_words = ["o", "e", "um", "da", "de", "à", "na", "foi"]
tokens = texto.lower().split(" ")
tokens_limpos = []
for token in tokens:
    if token not in stop_words:
        tokens_limpos.append( token )
texto = " ".join(tokens_limpos)
```

```python
wordcloud = WordCloud(width=1280, height=800, background_color='black').generate(texto)
```

```python
plt.figure(figsize=(15, 8))
plt.imshow( wordcloud )
plt.axis('off')
plt.show()
```
