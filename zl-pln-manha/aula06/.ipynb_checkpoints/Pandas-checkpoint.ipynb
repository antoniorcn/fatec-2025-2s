```python
stop_words = ["a", "o", "e", "para", "na", "no", "com", "de", "da", "do", "um", "uma"] 
```

```python
def limpar_texto( texto : str ) -> List[ str ]:
    texto_minusculo = texto.lower()
    tokens = texto_minusculo.split(" ")
    tokens_limpos = []
    for token in tokens:
        if token not in stop_words:
            tokens_limpos.append( token )
    return tokens_limpos
```

```python
limpar_texto(df_maior["paragrafo"][0])
```

```python
df_maior["tokens_limpos"] = df_maior["paragrafo"].apply( limpar_texto )
```
